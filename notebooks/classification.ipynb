```python
import pathlib
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

from superstore_analysis.pipelines.training_pipeline import ClassificationPipeline
from superstore_analysis.pipelines.feature_eng_pipeline import FeatureEng
from superstore_analysis.datasets import DataLoader
```

```python
script_dir = pathlib.Path(__name__).parent.resolve()
data_path = script_dir.parents[0] / 'data/Clustered.csv'
data = DataLoader().from_local(data_path)
data = data.drop(columns=['pca_1', 'pca_2'])
```

```python
data_eng = FeatureEng().fit_transform(data)
```

```python
data_eng.columns
```

```python
classifier_pipe = ClassificationPipeline()
train_set, test_set = classifier_pipe.preprocessing(
    data_eng, target='clusters')
classifier_pipe.train(train_set[0], train_set[1])
```

```python
classifier_pipe.evaluate(train_set=train_set, test_set=test_set)
```
